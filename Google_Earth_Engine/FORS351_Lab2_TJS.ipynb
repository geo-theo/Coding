***TASK 14***

```python
dataset = ee.ImageCollection('LANDSAT/LC08/C02/T1_L2').filterDate(
    '2021-07-01', '2021-08-01'
) #this code chooses our image and the date range

# Applies scaling factors.
def apply_scale_factors(image):
  optical_bands = image.select('SR_B.').multiply(0.0000275).add(-0.2)
  thermal_bands = image.select('ST_B.*').multiply(0.00341802).add(149.0)
  return image.addBands(optical_bands, None, True).addBands(
      thermal_bands, None, True
  )

#  Satellite data is stored as scaled integers, not real physical values.
#     Without this step, pixel values would not be physically meaningful.
# Optical bands (SR_B*) are converted to unitless surface reflectance
#     Thermal bands (ST_B*) are converted to temperature in Kelvin
# We used a function because the dataset contains many images. If we applied
#     this to every image the code would get very redundant.

dataset_scaled = dataset.map(apply_scale_factors)

#you will now visualize the data in RGB
visualization = {
    'bands': ['SR_B4', 'SR_B3', 'SR_B2'],
    'min': 0.0,
    'max': 0.3,
}

m = geemap.Map()
m.set_center(-113.9940, 46.8721, 8)
m.add_layer(dataset_scaled, visualization, 'True Color (432)')
m
```

***TASK 15***

```python
# Preprocessing function
def preprocess_landsat8(image):
    # QA masks
    qa_mask = image.select('QA_PIXEL').bitwiseAnd(int('11111', 2)).eq(0)
    # Each 1 in the above line corresponds to each bitmask value (i.e., Fill,
    # Dilated Cloud, Cirrius (high confidence), Cloud, and Cloud Shadow). The 2
    # tells us to interpret this as a base-2 (binary) number. The eq(0) indicates
    # we only want to include the highest value data.
    saturation_mask = image.select('QA_RADSAT').eq(0)

    # Apply scaling factors
    optical = image.select('SR_B.*').multiply(0.0000275).add(-0.2)
    thermal = image.select('ST_B.*').multiply(0.00341802).add(149.0)

    return (
        image
        .addBands(optical, None, True)
        .addBands(thermal, None, True)
        .updateMask(qa_mask)
        .updateMask(saturation_mask)
    )

dataset_filtered = dataset.map(preprocess_landsat8)

m = geemap.Map()
m.set_center(-113.9940, 46.8721, 8)
m.add_layer(dataset_filtered, visualization, 'True Color (432)')
m
```

***TASK 16***

```python
# Define the region of interest
# Example: Missoula Region
region = ee.Geometry.Rectangle([
    -114.35,  # West
    46.70,    # South
    -113.75,  # East
    47.05     # North
])

mso = dataset_filtered.mean().clip(region)

m = geemap.Map()
m.add_layer(mso, visualization, 'True Color (432)')
m
```

```python
# Define the region of interest
region = ee.Geometry.Rectangle([
    -114.35,  # West
    46.70,    # South
    -113.75,  # East
    47.05     # North
])

mso = dataset_filtered.mean().clip(region)

m = geemap.Map()
m.add_layer(mso, visualization, 'True Color (432)')
m.addLayer(region, {}, 'Region')
m.centerObject(region, 10)
m
```

***TASK 17***

```python
ndvi_expression = "(NIR - Red) / (NIR + Red)"

#mso is the name of your map you created from the ROI above!
ndvi = mso.expression(ndvi_expression,{
    'NIR': mso.select('SR_B5'),
    'Red': mso.select('SR_B4')})


# Import the Image function from the IPython.display module.
from IPython.display import Image

#note you can play around with the visualization if you like
ndvi_vis_params = {
    'min': 0,
    'max': 1,
    'dimensions': 1000,
    'region': region,
    'palette': ['A50026', 'D73027', 'F46D43', 'FDAE61', 'FEE08B',
                'FFFFBF','D9EF8B','A6D96A','66BD63','1A9850']
}

# Save as a variable and display the image.
ndvi_url = ndvi.getThumbURL(ndvi_vis_params)

display(Image(url=ndvi_url))
```

1. What do the greens and oranges/reds represent in this image?
> The greens represent areas with higher NDVI values, meaning healthier or denser vegetation. The oranges/reds represent lower NDVI values, meaning little vegetation, sparse vegetation, bare ground, urban surfaces, or possibly water.

2. What features have the most noticable high or low NDVI values?
> Dense, healthy vegetation has the highest NDVI values. Areas with little or no vegetation, such as bare soil, developed land, rock, or water, have the lowest NDVI values.

3. How would you expect this NDVI image to change in the winter?
> In winter, I would expect NDVI values to be lower because there is usually less active vegetation. Deciduous plants lose their leaves, grasses may be dormant, and snow cover can further reduce the appearance of healthy vegetation.
