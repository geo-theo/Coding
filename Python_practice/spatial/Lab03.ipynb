## Problem 3.3

**How many kilometers would you have to travel to visit every "Fish Creek" in Montana?** Don't worry about whether the route is the shortest possible route, just calculate the distance along *any* route that will bring you to *every* Fish Creek, even if you're visiting them in a random, zig-zagging order. You'll need to use the whole `geo` table for this...

That is, you should iterate over `geo.to_records()`, which allows you to refer to a column by number (position) *or* by name:

````python
for row in geo.to_records():
    name = row[2]
    name = row['name'] # Same result
    lat = row[3]
    lat = row['latitude'] # Same result

print(name) # Just one example
````

**The idea is to start at one Fish Creek (any one) and travel to the next, then the next, and so on.** The order in which you visit Fish Creeks doesn't matter.

Also, remember the `haversine()` function:

````python
import numpy as np

def haversine(p1, p2):
    x1, y1 = map(np.deg2rad, p1)
    x2, y2 = map(np.deg2rad, p2)
    dphi = np.abs(y2 - y1) # Difference in latitude
    dlambda = np.abs(x2 - x1) # Difference in longitude
    angle = 2 * np.arcsin(np.sqrt(np.add(
        np.power(np.sin(dphi / 2), 2),
        np.cos(y1) * np.cos(y2) * np.power(np.sin(dlambda / 2), 2)
    )))
    return float(angle * 6371e3) # Earth's radius: 6371e3 meters
````

**And remember that `haversine()` takes two sequences as arguments;**

```py
p1 = (some_longitude, some_latitude)
p2 = (another_longitude, another_latitude)
```

### Solution 3.3

````python
def fishcreek_travel_dist(df):
    # save list of all the Fish Creek coordinates in the order they appear
    fish_creeks = []
    for row in df.to_records():
        if row['name'] == "Fish Creek":
            lon = row['longitude']
            lat = row['latitude']
            fish_creeks.append((lon, lat))
    # use haversine function to measure distance between Fish Creek X coordinates and Fish Creek Y coordinates in meters, and keep adding each computed distance between two Fish Creek places to the total tally sum
    distance = 0
    for i in range(len(fish_creeks) - 1):
        p1 = fish_creeks[i]
        p2 = fish_creeks[i + 1]
        distance += haversine(p1, p2)
    # return result in km
    return distance / 1000
````

````python
total_km_traveled = fishcreek_travel_dist(geo)
print("Total kilometers traveled between all Fish Creek places: " + str(total_km_traveled) + "km")
````

### Solution 3.4

````python
# use function from previous lab to get decimal degrees, adding strip method to the hemisphere column to S/W as single character and make negative if need be
def dms_to_dd(deg, minutes, hemi):
    dd = float(deg) + float(minutes) / 60.0
    hemi = str(hemi).strip().upper()
    if hemi in ("S", "W"):
        dd = -dd
    return dd

# define function to iterate and read through columns and: (1) only keep rows with flag values 1 and 5; (2) convert minute coordinates to decimal degrees; (3) save row columns as new list elemnt
def build_clean_rows(co2):
    rows_out = []
    for row in co2.to_records():
        #flag = row['flag']
        if row['flag'] not in (1, 5):
            continue
        lat = dms_to_dd(row['lat_degrees'], row['lat_minutes'], row['lat_ns'])
        lon = dms_to_dd(row['lng_degrees'], row['lng_minutes'], row['lng_ew'])

        rows_out.append([row['flight'],row['date'], row['time'], lat, lon, row['pressure_hPa'], row['co2_ppm']])

    return rows_out
````

````python
# run function and save returns
clean_data = build_clean_rows(co2)

# show a single flagged row
print("Sample row values: " + str(clean_data[0]))

# count number of filtered rows (should be 919)
print("Number of rows after filtering:", len(clean_data)) 

# show flagged rows in dataframe
df = pd.DataFrame(clean_data)
df.columns = ['Flight', 'Date', 'Time', 'Lat', 'Lng', 'pressure_hPa', 'co2_ppm']
df
````
